```python
#Import package
import pandas as pd
import numpy as np
import ast
import matplotlib as mlt
import seaborn as sns
```

```python
df = pd.read_csv('imdb_data (1).csv')
```

```python
df.head()
```

```python
df['profit'] = df['revenue'] - df['budget']
df['ROI'] = df['profit'] / df['budget'].replace(0, np.nan)
def unpack_column(col):
  return col.fillna('[]').apply(ast.literal_eval)

```

```python
df['crew_list'] = unpack_column(df['crew'])
df['cast_list'] = unpack_column(df['cast'])
df['genres_list'] = unpack_column(df['genres'])

df['directors'] = df['crew_list'].apply(lambda x: [c['name'] for c in x if c['job'] == 'Director'])
df['producers'] = df['crew_list'].apply(lambda x: [c['name'] for c in x if c['job'] == 'Producer'])
df['actors'] = df['cast_list'].apply(lambda x: [c['name'] for c in x])
df['genre_names'] = df['genres_list'].apply(lambda x: [g['name'] for g in x])

print("Setup Complete! Dataset is ready for analysis.")
```

```python

max_profit_row = df['profit'].idxmax()


top_movie = df.loc[max_profit_row]

print("--- HIGHEST PROFIT MOVIE ---")
print(f"Title: {top_movie['title']}")
print(f"Profit: ${top_movie['profit']:,.2f}")
print(f"Directors: {top_movie['directors']}")
print(f"Producers: {top_movie['producers']}")
print(f"Top 5 Actors: {top_movie['actors'][:5]}")
```

```python

roi_by_lang = df.dropna(subset=['ROI']).groupby('original_language')['ROI'].mean().sort_values(ascending=False)

print("--- LANGUAGE WITH HIGHEST AVERAGE ROI ---")
print(f"Language Code: '{roi_by_lang.index[0]}'")
print(f"Average ROI: {roi_by_lang.iloc[0]:.2f}")
```

```python

unique_genres = df['genre_names'].explode().dropna().unique()

print("--- UNIQUE GENRES IN DATASET ---")

print(sorted(list(unique_genres)))
```

```python

prod_dir_table = df[['title', 'producers', 'directors']]
print("--- PRODUCERS & DIRECTORS TABLE ---")
display(prod_dir_table.head())
df_producers = df.explode('producers').dropna(subset=['producers', 'ROI'])


prod_stats = df_producers.groupby('producers').agg(
    Avg_ROI=('ROI', 'mean'),
    Movie_Count=('title', 'count')
)

top_3_producers = prod_stats[prod_stats['Movie_Count'] >= 3].sort_values(by='Avg_ROI', ascending=False).head(3)

print("\n--- TOP 3 PRODUCERS BY AVG ROI (Min 3 Movies) ---")
display(top_3_producers)
```

```python

roi_by_lang = df.dropna(subset=['ROI']).groupby('original_language')['ROI'].mean().sort_values(ascending=False)

print("--- LANGUAGE WITH HIGHEST AVERAGE ROI ---")
print(f"Language Code: '{roi_by_lang.index[0]}'")
print(f"Average ROI: {roi_by_lang.iloc[0]:.2f}")
```

```python
# Question 4
prod_dir_table = df[['title', 'producers', 'directors']]
print("--- PRODUCERS & DIRECTORS TABLE ---")
display(prod_dir_table.head())
df_producers = df.explode('producers').dropna(subset=['producers', 'ROI'])

prod_stats = df_producers.groupby('producers').agg(
    Avg_ROI=('ROI', 'mean'),
    Movie_Count=('title', 'count')
)


top_3_producers = prod_stats[prod_stats['Movie_Count'] >= 3].sort_values(by='Avg_ROI', ascending=False).head(3)

print("\n--- TOP 3 PRODUCERS BY AVG ROI (Min 3 Movies) ---")
display(top_3_producers)
```

```python
#question 5
df_actors = df.explode('actors').dropna(subset=['actors'])

most_frequent_actor = df_actors['actors'].value_counts().index[0]
movie_count = df_actors['actors'].value_counts().iloc[0]

print(f"--- ACTOR WITH MOST MOVIES ---")
print(f"Actor: {most_frequent_actor} ({movie_count} movies)")


actor_movies = df_actors[df_actors['actors'] == most_frequent_actor]

deep_dive_table = actor_movies[['title', 'genre_names', 'profit']].sort_values(by='profit', ascending=False)

print(f"\n--- DEEP DIVE: {most_frequent_actor.upper()}'S TOP 5 MOVIES BY PROFIT ---")
display(deep_dive_table.head(5))
```

```python
#question 6
df_directors = df.explode('directors').dropna(subset=['directors'])
top_3_directors_list = df_directors['directors'].value_counts().head(3).index.tolist()

print("--- TOP 3 DIRECTORS AND THEIR PREFERRED ACTORS ---")


for director in top_3_directors_list:

    director_movies = df_directors[df_directors['directors'] == director]

    director_actors = director_movies.explode('actors').dropna(subset=['actors'])

    favorite_actor = director_actors['actors'].value_counts().index[0]
    times_cast = director_actors['actors'].value_counts().iloc[0]

    print(f"Director: {director} | Most Cast Actor: {favorite_actor} ({times_cast} times)")
```
